```python
data = pd.read_csv("employee.csv")
data
```

```python
data.filter(["name", "department"])
```

```python
data.filter(["name", "department"], axis = 0)
```

```python
data.filter(["name", "department"], axis = 1)
```

```python
data[data.department == "Sales"]
```

```python
data[data.department.isin(["Sales", "Finance"])] # memanggil data lebih dari 1
```

```python
data[data.performance_score >= 700]
```

```python
data["age"] = data.age.fillna(data.age.mean())
data
```

```python
data["income"] = data.income.fillna(data.income.median()) # nilai tengah
data
```

```python
upper = data.performance_score.mean() + 3 * data.performance_score.std()
lower = data.performance_score.mean() - 3 * data.performance_score.std()

data = data[(data.performance_score < upper) & (data.performance_score > lower)]
data
```

```python
upper = data.performance_score.quantile(.99)
lower = data.performance_score.quantile(.01)

data = data[(data.performance_score < upper) & (data.performance_score > lower)]
data
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

data["baru"] = scaler.fit_transform(data[["performance_score"]])

data.head()
```

```python
data["first name"] = data.name.str.split(" ").map(lambda var: var[0])
data["last name"] = data.name.str.split(" ").map(lambda var: var[1])

data
```
