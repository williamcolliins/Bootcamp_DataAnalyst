PENDEKATAN PROSEDURAL

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("penyakit_jantung.csv")
data
```

```python
trestbps = data.groupby("age")["trestbps"].sum()
chol = data.groupby("age")["chol"].sum()

sns.lineplot(x = trestbps.index, y = trestbps.values, label = "trestbps")
sns.lineplot(x = chol.index, y = chol.values, label = "chol")
plt.legend()
```

```python
plt.subplot(121) # (baris, kolom, letak plot)
sns.lineplot(x = trestbps.index, y = trestbps.values)
plt.xlabel("Age")
plt.ylabel("Trestbps")

plt.subplot(122)
sns.lineplot(x = chol.index, y = chol.values, color = "red")
plt.xlabel("Age")
plt.ylabel("Chol")

plt.tight_layout()
plt.show
```

```python
plt.subplot(121) # (baris, kolom, letak plot)
sns.lineplot(x = trestbps.index, y = trestbps.values)
plt.xlabel("Age")
plt.ylabel("Trestbps")

plt.subplot(122)
sns.scatterplot(x = chol.index, y = chol.values, color = "red")
plt.xlabel("Age")
plt.ylabel("Chol")

plt.tight_layout()
plt.show
```

PENDEKATAN OBJECT ORIENTED

```python
fig = plt.figure()
axes = fig.add_axes([0,0,0.7,0.9]) # (left, bottom, width, height)
axes.plot(trestbps.index, trestbps.values)
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1.3,1.2]) # (left, bottom, width, height)
axes1.plot(trestbps.index, trestbps.values)
axes1.set_xlabel("Age")
axes1.set_ylabel("Trestbps")

axes2 = fig.add_axes([0.2,0.8,0.3,0.3])
axes2.plot(chol.index, chol.values, color = "green")
axes2.set_xlabel("Age")
axes2.set_ylabel("Chol")
```

```python
fig, axes = plt.subplots(2,2) # (baris,kolom)

# baris 1 kolom 1
axes[0][0].plot(trestbps.index, trestbps.values, color="red")

# baris 1 kolom 2
axes[0][1].plot(trestbps.index, trestbps.values, color="blue")

# baris 2 kolom 1
axes[1][0].plot(trestbps.index, trestbps.values, color="green")

# baris 2 kolom 2
axes[1][1].plot(trestbps.index, trestbps.values, color="purple")

fig.tight_layout()
```

```python
region = dataSS.groupby('Region')['Sales'].sum()

print(f"sales tertinggi : {region.idxmax()}")
print(f"sales terendah : {region.idxmin()}")
sns.barplot(y = region.index, x = region.values)
```

4

```python
profitsales_percategory = dataSS.groupby('Category')[['Profit', 'Sales']].sum()
profitsales_percategory
```

```python
profitsales_percategory.index
```

```python
profitsales_percategory.values
```

```python
sns.lineplot(x = profitsales_percategory.index, y = profitsales_percategory.values[0: 0], label = 'Profit')
sns.linelot(x = profitsales_percategory.index, y = profitsales_percategory.values[0: 1], label = 'Sales')
plt.title('Profit And Sales Per Category')
plt.legend()
plt.show()
```

```python
print(f"Kategori hasil tertinggi: \n{profitsales_percategory.idxmax()}")
print(f"Kategori hasil terendah: \n{profitsales_percategory.idxmin()}")
```
