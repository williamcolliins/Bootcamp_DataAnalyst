5. Buat scatter plot dengan income pada sumbu X dan life_expec pada sumbu Y.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.scatter(df['income'], df['life_expec'])

plt.title('Scatter Plot: Income vs Life Expectancy')
plt.xlabel('Income')
plt.ylabel('Life Expectancy')

plt.grid(True)
plt.show()
```

6. Tampilkan jumlah gdp setiap negara menggunakan barplot, serta tampilkan negara dengan jumlah gdp terbanyak dan terendah.

```python
import seaborn as sns
import matplotlib.pyplot as plt

def visualisasi_gdp_sederhana(df):
    negara_tertinggi = df.loc[df['gdpp'].idxmax(), 'country']
    gdp_tertinggi = df['gdpp'].max()
    negara_terendah = df.loc[df['gdpp'].idxmin(), 'country']
    gdp_terendah = df['gdpp'].min()

    print(f"Negara dengan GDP tertinggi: {negara_tertinggi} ({gdp_tertinggi})")
    print(f"Negara dengan GDP terendah: {negara_terendah} ({gdp_terendah})")

    plt.figure(figsize=(12, 6))
    sns.barplot(data=df, x='country', y='gdpp', color='blue')
    plt.title('Barplot GDP per Country', fontsize=14)
    plt.xlabel('Country', fontsize=12)
    plt.ylabel('GDP (gdpp)', fontsize=12)
    plt.xticks(rotation=90, fontsize=8)
    plt.tight_layout()
    plt.show()

visualisasi_gdp_sederhana(df)
```

7. Tampilkan inflasi negara Afghanistan, albania dan angola menggunakan lineplot.

```python
import seaborn as sns
import matplotlib.pyplot as plt


negara = df[df['country'].isin(['Afghanistan', 'Albania', 'Angola'])]

plt.figure(figsize=(8, 6))
sns.lineplot(data=negara, x='country', y='inflation', marker='o', sort=False)

plt.title('Inflasi Negara: Afghanistan, Albania, dan Angola')
plt.xlabel('Country')
plt.ylabel('Inflation')
plt.xticks(fontsize=10)
plt.grid(True)

plt.tight_layout()
plt.show()
```
