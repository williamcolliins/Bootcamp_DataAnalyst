```python
# pastikan java sudah terinstal di vsc untuk membaca file pdf dan install tabula-py (library tabula)

from tabula import read_pdf
import pandas as pd

pdf_file = "bri.pdf"

# membaca tabel dari halaman pertama
df = read_pdf(pdf_file, pages=1, lattice=True)
print(df)
```

```python
# mengekstraksi banya tabel
# membaca semua tabel di halaman
tables = read_pdf(pdf_file, pages="all", multiple_tables=True, lattice=True)

# menampilkan jumlah tabel yang ditemukan
print(f"Ditemukan {len(tables)} tabel.")

# menamipilkan tabel pertama
print(tables[0])
```

```python
# menyimpan tabel ke csv
for i, table in enumerate(tables):
    csv_file = f"table_{i + 1}.csv"  # Nama file CSV untuk setiap tabel
    table.to_csv(csv_file, index=False)
    print(f"Tabel {i + 1} disimpan sebagai {csv_file}")
```

```python
from tabula import read_pdf
import pandas as pd

pdf_file = "bri.pdf"

# ekstraksi data tabel
tables = read_pdf(pdf_file, pages="all", multiple_tables=True, lattice=True)

# menampilkan jumlah tabel yang ditemukan
print(f"Terdapat {len(tables)} tabel yang diekstrak.")

# ambil tabel aset (tabel pertama)
table_aset = tables[0]

# ambil tabel rasio kinerja (tabel kedua)
table_rasio = tables[1]

print(table_aset)
print(table_rasio)
```

```python
# cek struktur tabel aset
print(table_aset.head())  # menampilkan beberapa baris pertama
print(table_aset.columns)  # menampilkan nama kolom
print(len(table_aset.columns))  # menampilkan jumlah kolom
```

```python
# hapus kolom tambahan jika ada
table_aset = table_aset.iloc[:, :5]  # hanya pilih 5 kolom pertama
```

```python
# atur nama kolom
table_aset.columns = ["POS", "Individual 30 June 2023", "Individual 31 December 2022", "Consolidated 30 June 2023", "Consolidated 31 December 2022"]
```

```python
# cek dataframe
print(table_aset.info())  # cek tipe data dan kolom
print(table_aset.head())  # cek data awal
```
